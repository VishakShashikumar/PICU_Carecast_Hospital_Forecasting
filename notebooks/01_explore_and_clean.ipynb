```python
import pandas as pd

df = pd.read_csv("hospital_capacity.csv")

# Convert date column
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Filter only Massachusetts
df_ma = df[df['state'] == 'MA'].copy()

df_ma.head()
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')

df_ma = df[df['state'] == 'MA'].copy()
df_ma.head()
```

```python
daily_ma = (
    df_ma.groupby('date', as_index=False)[[
        'staffed_pediatric_icu_bed_occupancy',
        'total_staffed_pediatric_icu_beds'
    ]].sum()
)

daily_ma.head()
```

```python
daily_ma['staffed_pediatric_icu_bed_occupancy'] = pd.to_numeric(
    daily_ma['staffed_pediatric_icu_bed_occupancy'], errors='coerce'
)

daily_ma['total_staffed_pediatric_icu_beds'] = pd.to_numeric(
    daily_ma['total_staffed_pediatric_icu_beds'], errors='coerce'
)
```

```python
daily_ma = daily_ma[daily_ma['total_staffed_pediatric_icu_beds'] > 0].copy()
```

```python
daily_ma.dtypes
```

```python
daily_ma['picu_utilization_pct'] = (
    daily_ma['staffed_pediatric_icu_bed_occupancy'] /
    daily_ma['total_staffed_pediatric_icu_beds'] * 100
)

daily_ma[['date', 'staffed_pediatric_icu_bed_occupancy',
          'total_staffed_pediatric_icu_beds',
          'picu_utilization_pct']].head()
```

```python
daily_ma = daily_ma.sort_values('date')
daily_ma.tail()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
plt.plot(daily_ma['date'], daily_ma['picu_utilization_pct'])
plt.title('Pediatric ICU Bed Utilization (%) - Massachusetts')
plt.xlabel('Date')
plt.ylabel('Utilization %')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df_prophet = daily_ma[['date', 'picu_utilization_pct']].rename(
    columns={'date': 'ds', 'picu_utilization_pct': 'y'}
)

df_prophet.head()
```

```python
# Just to be sure this exists:
df_prophet = daily_ma[['date', 'picu_utilization_pct']].rename(
    columns={'date': 'ds', 'picu_utilization_pct': 'y'}
)
df_prophet.head()
```

```python
train = df_prophet.iloc[:-30].copy()
test = df_prophet.iloc[-30:].copy()

print(train.tail())
print(test.head())
```

```python
import statsmodels.api as sm
```

```python
# Set date as index for statsmodels
train_ts = train.set_index('ds')['y']

model = sm.tsa.SARIMAX(train_ts, order=(1, 1, 1), seasonal_order=(1, 0, 1, 7))
results = model.fit(disp=False)
results.summary()
```

```python
# Number of forecast steps = length of test set
n_forecast = len(test)

# Generate predictions
forecast_values = results.forecast(steps=n_forecast)

forecast_values.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))

# Training data
plt.plot(train['ds'], train['y'], label='Train', linewidth=2)

# Test (actual)
plt.plot(test['ds'], test['y'], label='Test (Actual)', linewidth=2)

# Forecasted values
plt.plot(test['ds'], forecast_values.values, label='Forecast', linestyle='--', linewidth=2)

plt.title("SARIMAX Forecast — Pediatric ICU Utilization (MA)")
plt.xlabel("Date")
plt.ylabel("Utilization %")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

y_true = test['y'].values
y_pred = forecast_values.values

mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))

mae, rmse
```

```python
pred = results.get_forecast(steps=n_forecast)
pred_ci = pred.conf_int()

plt.figure(figsize=(12,5))

plt.plot(train['ds'], train['y'], label='Train')
plt.plot(test['ds'], test['y'], label='Actual')
plt.plot(test['ds'], pred.predicted_mean, label='Forecast', linestyle='--')

plt.fill_between(
    test['ds'],
    pred_ci.iloc[:, 0],
    pred_ci.iloc[:, 1],
    color='lightgray',
    alpha=0.4
)

plt.title("SARIMAX 30-Day Forecast with Confidence Interval")
plt.xlabel("Date")
plt.ylabel("Utilization %")
plt.legend()
plt.grid(True)
plt.show()
```
